# STEP ONE

Running with HiCFoundation model

- Enhance .pairs
- convert enhanced results to .txt

```python

import os
import subprocess
import sys

pairs_dir = "/storage/ireneky/ireneky_cse428/brain_hires" #brain_10
output_dir = "/storage/ireneky/ireneky_cse428/brain_hires_enhanced" #brain_10_enhanced
model_path = "/storage/ireneky/HiCFoundation/hicfoundation_model/hicfoundation_schic.pth.tar"
inference_script = "/storage/ireneky/HiCFoundation/inference.py"

os.makedirs(output_dir, exist_ok=True)

pairs_files = [f for f in os.listdir(pairs_dir) if f.endswith(".pairs")]

for f in pairs_files:
    input_path = os.path.join(pairs_dir, f)
    output_file = os.path.splitext(f)[0] + "_enhanced"
    output_path = os.path.join(output_dir, output_file)
    
    print(f"Enhancing: {f} → {output_file}")
    
    cmd = [
        sys.executable, inference_script,
        "--input", input_path,
        "--batch_size", "4",
        "--resolution", "1000000",
        "--task", "5",
        "--input_row_size", "224",
        "--input_col_size", "224",
        "--stride", "20",
        "--bound", "250",
        "--model_path", model_path,
        "--output", output_path,
        "--gpu", "0,1"
    ]
    
    subprocess.run(cmd)

```

# STEP TWO

Convert results to Higashi Format

```python
import os
import pickle
import pandas as pd

# === Configuration ===
base_dir = "/storage/ireneky/ireneky_cse428/brain_hires_enhanced" #brain_10_enhanced
output_dir = "/storage/ireneky/ireneky_cse428/higashi_data_v1"
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, "data.txt")
resolution = 1000000

def matrix_dict_to_df(matrix_dict, cell_id, resolution):
    rows = []
    for chrom, mat in matrix_dict.items():
        for i in range(mat.shape[0]):
            for j in range(i, mat.shape[1]):
                count = mat[i, j]
                if count > 0:
                    pos1 = i * resolution
                    pos2 = j * resolution
                    rows.append([cell_id, chrom, pos1, chrom, pos2, count])
    return pd.DataFrame(rows, columns=["cell_id", "chrom1", "pos1", "chrom2", "pos2", "count"])

# === Loop over subfolders and collect .pkl outputs ===
all_rows = []
subfolders = sorted(os.listdir(base_dir))

for cell_id, folder in enumerate(subfolders):
    pkl_path = os.path.join(base_dir, folder, "HiCFoundation_sc_enhanced.pkl")
    if not os.path.exists(pkl_path):
        print(f"Skipping {folder}: no HiCFoundation_sc_enhanced.pkl")
        continue
    with open(pkl_path, "rb") as f:
        data = pickle.load(f)
    df = matrix_dict_to_df(data, cell_id, resolution)
    all_rows.append(df)
    print(f"Processed cell {cell_id}: {folder}")

# === Save merged data.txt ===
merged_df = pd.concat(all_rows, ignore_index=True)
merged_df.to_csv(output_path, sep="\t", index=False)
print(f"\n Saved: {output_path} with {len(merged_df)} rows from {len(all_rows)} cells")
```

```python
import pandas as pd
import pickle
import os

# === Load metadata ===
meta_path = "/storage/ireneky/ireneky_cse428/brain_hires_meta.csv"  # update if using .csv #brain_meta.csv
df = pd.read_csv(meta_path, sep=",")

# Optional: check it loaded correctly
#assert df.shape[0] == 10, "Expected 10 rows"
#assert "Cell type" in df.columns or "Celltype" in df.columns, "Missing cell type column"

# === Rename columns for clarity ===
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

# === Extract Higashi-compatible label_info ===
label_info = {
    "cell type": df["cell_type"].tolist(),
    "coverage": df["dedup_contacts"].tolist(),
    "batch": ["batch_1"] * len(df)  # optional: make smarter if you have batches
}

# === Save ===
output_path = "/storage/ireneky/ireneky_cse428/higashi_data_v1/label_info.pickle"
with open(output_path, "wb") as f:
    pickle.dump(label_info, f)

print(f" Saved label_info.pickle with {len(df)} entries to: {output_path}")
```
